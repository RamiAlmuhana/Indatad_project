# **Omgaan met Missende Waarden**
In deze stap identificeer ik missende waarden in de dataset en neem ik beslissingen over hoe ik deze zal behandelen. Ik vul missende waarden in de kolommen 'likes' en 'comment_count' in met de gemiddelde, mediaan, of meest voorkomende waarde (modus).

Ik zoek naar verschillende manieren waarop missende waarden kunnen worden weergegeven, zoals NaN, None of lege strings.

```python
# Check of er verschillende formaten voor missende waarden zijn
missende_waarden = df.isin([np.nan, None, 'NA', 'null', '']).sum()
print(missende_waarden)
```

Ik verwijder rijen met missende waarden in de dataset.

```python
# Verwijderen van missende waarden:
df.dropna(inplace=True)  # Verwijder rijen met missende waarden
```

Ik vul missende waarden in de kolom 'likes' met het gemiddelde, de mediaan of de meest voorkomende waarde.

Ik vul missende waarden in de kolom 'comment_count' op dezelfde manier in.



```python
# Mean
mean_likes = df['likes'].mean()
df['likes'] = df['likes'].fillna(mean_likes)

# Median
median_likes = df['likes'].median()
df['likes'] = df['likes'].fillna(median_likes)

# Mode
mode_likes = df['likes'].mode()[0]
df['likes'] = df['likes'].fillna(mode_likes)

# Mean
mean_comment_count = df['comment_count'].mean()
df['comment_count'] = df['comment_count'].fillna(mean_comment_count)

# Median
median_comment_count = df['comment_count'].median()
df['comment_count'] = df['comment_count'].fillna(median_comment_count)

# Mode
mode_comment_count = df['comment_count'].mode()[0]
df['comment_count'] = df['comment_count'].fillna(mode_comment_count)

df.head()
```

# **Feature Engineering**
In deze stap voeg ik nieuwe kolommen toe aan de dataset. Zo voeg ik bijvoorbeeld de lengte van de titel toe, bereken ik de verhouding van likes tot views, en maak ik de kolom 'days_since_published'. Daarnaast verwijder ik enkele kolommen die ik niet nodig heb, zoals 'dislikes' en 'duration' in ISO 8601-formaat.


```python
def convert_duration_to_seconds(duration):
    if pd.isnull(duration) or duration == '':
        return 0

    minutes = 0
    seconds = 0

    duration = duration.replace('PT', '').replace('S', '')

    if 'H' in duration:
        hours, duration = duration.split('H')
        minutes += int(hours) * 60

    if 'M' in duration:
        minutes_part, duration = duration.split('M')
        minutes += int(minutes_part)

    if duration:
        seconds = int(duration)

    return minutes * 60 + seconds

# Toepassen van de functie op de 'duration' kolom
df['duration_seconds'] = df['duration'].apply(convert_duration_to_seconds)

# Imputatie voor de 'duration_seconds' kolom
mean_duration = df['duration_seconds'].mean()
df['duration_seconds'] = df['duration_seconds'].fillna(mean_duration)

# Controleer de eerste paar rijen van de DataFrame
df.head()
```

Hier maak ik een niuewe kolom genaamd (days_since_published) door (published_at) te gebruiken.

```python
# Ensure 'published_at' is tz-naive (remove timezone info)
df['published_at'] = pd.to_datetime(df['published_at']).dt.tz_localize(None)

# Calculate 'days_since_published'
df['days_since_published'] = (pd.Timestamp.today() - df['published_at']).dt.days
df['days_since_published'] = df['days_since_published'].replace(0, 1)
df.head()
```

In deze stap zet ik de kolom 'published_at' om naar drie afzonderlijke kolommen: 'published_Year', 'published_Month', en 'published_Day'. Dit maakt het makkelijker om later analyses te doen op deze tijdsgegevens.

```python
df['published_at'] = pd.to_datetime(df['published_at'])
df['published_Year'] = df['published_at'].dt.year
df['published_Month'] = df['published_at'].dt.month
df['published_Day'] = df['published_at'].dt.day

df.head(4)
```
